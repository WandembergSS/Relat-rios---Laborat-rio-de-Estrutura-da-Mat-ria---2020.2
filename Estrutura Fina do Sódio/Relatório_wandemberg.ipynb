## 1 - Introdução
  
  O primeiro modelo atômico a prever corretamente as energias de emissão de um átomo, foi o modelo de Bohr, formulado em 1913 por Niels Bohr, em uma série de artigos. Após concluir seu Phd, ele trabalhou com J. J. Thomson e Ernest Rutherford, acompanhando seus trabalhos sobre a estrutura da matéria. Ao retornar para a Universidade de Copenhague, onde fez seu Phd, elaborou sua teoria para o átomo de hidrogênio, englobando também a teoria de Max Planck sobre o corpo negro e de Albert Einstein do efeito fotoelétrico. Baseado em órbitas circulares para o elétron, com momento angular quantizado, dando uma explicão teórica para a fórmula de Rydberg, elaborada a partir de resultados experimentais.
  
  Apesar do seu grande sucesso, o modelo de Bhor não foi suficiente para interpretar a emissão de átomos não hidrogenóides. O modelo de Sommerfeld para os metais alcalinos é uma alternativa que leva em conta a interação entre os elétrons. Considerando órbitas elípticas, diferente do modelo de Bohr, devemos agora considerar a quantização do momento linear, o que nos leva a dois números quênticos; o principal $n$ e o de momento angular $l$. Essas órbitas possuem uma diminuição em suas excentricidades a medida que o número l aumenta. Assim, órbitas com menores números de l terão maior penetração nas camadas e, por consequência, maior energia de ligação. Com isso, é então explicado as equações empíricas obtidas por Rydber em 1889, após generalizar sua equação para átomos hidrogenoides, introduzindo um defeito quântico. Esse defeito é funão do número $l$, gerando um número quântico principal efetivo na experessão para o comprimento de onda em transições:

$$   \frac{1}{\lambda} = R \left[ \frac{1}{(n_1 - \mu_1(l))^2} - \frac{1}{(n_2 - \mu_1(l))^2} \right]
$$


 Em relação ao espectro do hidrogênio, encontramos também outra diferença. Podemos perceber, ao analisiar o espectro de elementos alcalinos, que existe um desdobramento  dos níveis, chamada "Estrutura Fina" do átomo. Esse fenômeno ocorre devido aos elétrons possuírem um momento magnético intríseco. As primeiras evidências dessa propriedade da matéria foram observadas em um experimento dos físicos Otto Stern e Walther Gerlach. Ao passar um feixe de átomos metálicos neutros por uma região com um campo magnético não homogêneo, foi observado que o feixe incidente foi defletido em duas direções distintas. Primeiramente, os físicos interpretaram o resultado como comprovação da quantização do momento angular magnético devido ao movimento óbital, mas existia o problema de que era esperado três direções distintas de deflexão, uma para cada número quântico de momento angular possível para um estado $l=1$. Após isso, com a elaboaração da teoria do spin por Wolfgang Pauli, as consequências dessa propriedade da matéria se tornaram previsíveis.

## 3 - Procedimentos e Resultados

### 3.1 - Caracterização da Rede de Difração
 Para a medição do espaçamento $d$ entre as ranhuras da rede de difração, essa rede difratará o raios de um laser de Hélio-Neônio, de comprimento de onda $\lambda = 632,8 nm$. Medindo a separação entre os máximos de difração em um anteparo colocado a uma distância D da rede, os feixes de ordem ±1, e depois ±2, foram colocados nas extremidades de um anteparo de comprimento L = (15,0 ± 0,1) cm, respectivamente, obtemos:

  $D_{n=1} = (17,9 ± 0,1)cm$; e 
  
  $D_{n=2} = (6,2 ± 0,1)cm$.


fig xx.jpg

Fig. XX, Fundamentals of physics / Jearl Walker, David Halliday, Robert Resnick—10th edition.

 Onde foi utilizada uma régua com resolução 0,5 cm.  Assim, como 
$$ sin(\theta) = \frac{L}{\sqrt{L^2 + 4D^2}};$$ $$$$
temos:

 $$ d = \frac{n\lambda}{sin(\theta)} = \frac{n\lambda \sqrt{L^2 + 4D^2} }{L};$$

e
$$$$
$$\sigma_d = d \frac{(2D/L)^2}{1 + (2D/L)^2} \sqrt{ 
  \left( \frac{\sigma_L}{L}\right)^2 
   + \left(\frac{\sigma_D }{D}\right)^2  }.    $$
   $$$$
   Que nos leva a:



```python
df = pd.DataFrame()

df['n'] = range(1,3)
df['L(cm)'] = [15.0, 15.0]
df['sigmaL'] = [0.1, 0.1]
df['D(cm)'] = [17.9, 6.2]
df['sigmaD'] = [0.1, 0.1]

df['d(nm)'] = df['n'] * (632.8)* sqrt(df['L(cm)']**2 + 4*(df['D(cm)'])**2)/df['L(cm)']

df['sigmad(nm)'] = df['d(nm)']  *   ((((2*df['D(cm)'])/(df['L(cm)']))**2) / (1+ ((2*df['D(cm)'])/(df['L(cm)']))**2))  *  sqrt( (df['sigmaL'] / df['L(cm)'])**2  +  (df['sigmaD'] / df['D(cm)'])**2)

df
```

Assim, nossa estimativa para a a separação entre as ranhuras na rede de difração é

$$
d = (1640 ± 8)  nm.
$$
Onde a incerteza foi calculada por:
$$
\sigma_d = \frac{\sqrt{\sigma_{d(n=1)}^2 + \sigma_{d(n=2)}^2}}{2}.
$$
E a densidade de linhas será:
$$
\alpha = \frac{1}{d} = 610 \frac{linhas}{ mm}.

### 3.2 - Comprimentos de onda
Foram medidos ângulos para 4 linhas de primeira ordem de difração e 4 de dubletos distintos de segunda ordem. Para pontos suficientemente distantes da rede de difração, os raios chegarão aproximadamente paralelos em um anteparo que os intercepta em um ponto de máximo da difração. Assim, como nesse ponto a diferença de percurso entre os raios é um múltiplo do comprimento de onda, teremos que relação entre o ângulo de incidência $\theta$ e a distância **d** entre as ranhuras 

$$
sin(\theta) = \frac{n\lambda}{d},
$$

para uma incidência *normal* de raios sobre a rede.

#### 3.2.1 - Transições de primeira ordem
Da figura X, temos:

$\theta_1 = 21,83^o$ / $5^2 S_{1/2}$ --> $3^2 P_{1/2}$
$$$$

$\theta_2 = 20,77^o$ / $3^2 S_{3/2}$ --> $3^2 S_{1/2}$
$$$$

$\theta_3 = 20,57^o$ / $4^2 D_{5/2, 3/2}$ --> $3^2 P_{3/2}$
$$$$

$\theta_4 = 17,88^o$ / $5^2 D_{5/2, 3/2}$ --> $3^2 P_{3/2}$


```python
df = pd.DataFrame()

df['Theta(graus)'] = [(21.83*pi)/180, (20.77*pi)/180, (20.57*pi)/180, (17.88*pi)/180]
d = 1640
deld = 8
deltheta = (0.02*pi)/180
df['Lambda(angstron)'] = 10*d*sin(df['Theta(graus)'])
df['dLambda(angstron)'] = df['Lambda(angstron)']*sqrt((deld/d)**2 + (deltheta/tan(df['Theta(graus)']))**2 )


df
```

## 4 - Conclusão
Como esperávamos, as diferenças de energias dos dubletos são muito pequenas, comparado as transições que medimos em 3.2.1. Sem considerações relativísticas, apesar de serem importantes, podemos calcular a diferênça de energia em um desses desdobramentos. Na teoria de spin surge uma nova variável quântica chamada vetor de spin $S$, que tem suas componentes dadas pelo número quântico de spin $m_s$:

$$            S = \hbar m_s.
$$
Somos então levados a um momento magnético

$$ \mu_s = -g_s \mu_B \frac{S}{\hbar},
$$

onde $g_s$ é uma constante experimental e $\mu_B$ é o magnéton de Bohr.
 Como, no movimento orbital dos elétrons, é gerado um campo magnético interno, onde poderiamos pensar que existe um estado único, temos na verdade dois estados que possuem energias ligeiramente distintas.
